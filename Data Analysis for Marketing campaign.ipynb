**1-Convert type of this columns to string: `[ID , AcceptedCmp1,2,3,4,5 , Response , complain]`**

```python
columns =['ID','AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response','Complain']
df[columns] = df[columns].astype(str)
```

********
**2-Columns ` Dt_Customer` should convert to date type instance of string.**

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
print(df['Dt_Customer'].dtype)
```

****
**3-Column `Country` edite to contain name of country instance code of country.**

```python
# d =['Au','aus','GER','ind','id','ge','sp']
# countries = list(pycountry.countries)
def convert_country(i):
    try:
        return pycountry.countries.get(alpha_2=i[:2]).name
    except:
        return i
    
df['Country'] = df['Country'].apply(convert_country )
```

```python
df['Country'].head(20)
```

****
**4-Change nane of column ` Income ` to `Income` or use `strip` function with this column.**

```python
df.rename(columns={" Income ":"Income"}, inplace=True)
print(df.columns)
```

****
**5-remove `$` and `,` in `Income` column and convert this column to float.**

```python
def convert_to_float(x):
    try:
        return  pd.to_numeric(x.replace('$','').replace(',',''))
    except:
        if type(x)== float:
            return x
        else:
            return 'na'
        
        

df['Income'] = df['Income'].apply(lambda x:convert_to_float(x))
```

```python
df['Income'].head()
```

****
**6-Fill null values of `Income` column with the mean of this column.**

```python
round(df['Income'].mean(),2)
```

```python
df['Income'].fillna(value= round(df['Income'].mean(),2), inplace=True)
```

```python
pd.isnull(df['Income']).sum()
```
